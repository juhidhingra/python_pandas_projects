### Create a smaller dataframe with a subset of all features

```python
df_movies = pd.read_csv('movies_metadata_w3resource.csv')
df_movies.head()
```

```python
small_df = df_movies[['title', 'release_date', 'budget', 'revenue', 'runtime']]
print("Smaller DataFrame:")
print(small_df.head())
```

### Sort the DataFrame based on release_date

```python
df_movies.sort_values('release_date').head()
```

```python
small_df.sort_values('release_date').head()
```

### Access those movies, released after 1995-01-01

```python
small_df.head()
```

```python
result = small_df[small_df['release_date'] > '1995-01-01']
print("DataFrame based on release date>'1995-01-01':")
result.head()
```

### Sort movies on runtime in descending order

```python
result = small_df.sort_values('runtime',ascending=False)
print("DataFrame sort on Runtime")
result.head()
```
